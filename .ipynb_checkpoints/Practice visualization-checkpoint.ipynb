```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('2017.csv')
```

```python
df.columns
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(df['Happiness.Score'], bins=15, color='green')
plt.xlabel('Happiness Score (1-10)')
plt.title('Happiness Score')

plt.subplot(1,2,2)
plt.scatter(x=df['Happiness.Score'], y=df['Generosity'], color='purple')
plt.xlabel('Happiness Score')
plt.ylabel('Generosity')
plt.title('Does happiness correlate to generosity?')
plt.show()
```

What we can see from the above data is that the mode of the happiness score seems to be around 5, but a bigger takeaway is that there are more people above the happiness score of 5 than below. Although, it seems that happiness does not seem to correlate with the perceived generosity of one's own country.

```python
df[df['Happiness.Score']>=5].count() > df[df['Happiness.Score']<=5].count()
#so we know that there are more countries that are above the median happiness level than below. One could wager that
#if we had access to previous data, we might see a trend where either the mean happiness score is lower or more 
#people/countries are below the happiness score median
```

```python
plt.plot(df['Economy..GDP.per.Capita.'])
plt.xlabel('Happiness Rank')
plt.ylabel('Per Capita GDP')
plt.title('Happiness and the Economy')
plt.show()
```

The DataFrame is setup in a way that the index represents the happiness rank, so we can conclude that the happiness rank and per capita GDP are negatively correlated.

```python
plt.scatter(x=df['Happiness.Score'], y=df['Trust..Government.Corruption.'])
plt.xlabel('Happiness Score')
plt.ylabel('Perception of Corruption')
plt.show()
```

This is an interesting look at how our perception of corruption in the government contribute to a country's happiness score.
The y-label is scored from 0-1 with 0 being that perceived corruption in the governemnt does not contribute to a country's happiness while 1 being that it does. What we can infer from this data is that there is a weak, positive correlation between the happiness of a country and the understanding that corruption in the government can affect its citizens' happiness. 

```python
plt.scatter(y=df['Freedom'], x=df['Happiness.Rank'])
plt.title('How Important is Freedom in Measuring Happiness?')
plt.xlabel('Happiness Rank')
plt.ylabel('Freedom')
plt.show()
```

Lastly, we see analyze the extent to which freedom contributes to a country's happiness. As the happiness rank goes higher, countries start to appreciate the boons of freedom more and rate it as highly affecting their happiness while lower ranked countries seem to not hold freedom as high in regard.
